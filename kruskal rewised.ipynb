```python
import random
import numpy as np
class cell:
    def __init__(self,*pos):
        self.y ,self.x,self.id = pos
        self.ancestor = self
        self.neigs = []
    def connect(self,c):
        self.ancestor = c
        if abs(c.y-self.y) <= 1 and abs(c.x-self.x) <=1:
            return (c.y-self.y,c.x-self.x)
        else:
            return (-np.sign(c.y-self.y),-np.sign(c.x-self.x))
class maze:
    def __init__(self,w,h):
        self.map = []
        self.w , self.h = w,h
        self.cells = []
        self.ids = []
        for y in range(h):
            self.map.append([])
            self.ids.append([])
            for x in range(w):
                c = cell(y,x,w*y+x)
                self.map[y].append(c)
                self.cells.append(c)
                self.ids[y].append(c.id)
        for y in range(h):
            for x in range(w):
                if x != 0:
                    self.map[y][x].neigs.append(self.map[y][x-1])
                else:
                    self.map[y][x].neigs.append(self.map[y][w-1])
                if x != w-1:
                    self.map[y][x].neigs.append(self.map[y][x+1])
                else:
                    self.map[y][x].neigs.append(self.map[y][0])
                if y != 0:
                    self.map[y][x].neigs.append(self.map[y-1][x])
                else:
                    self.map[y][x].neigs.append(self.map[h-1][x])
                if y != h-1:
                    self.map[y][x].neigs.append(self.map[y+1][x])
                else:
                    self.map[y][x].neigs.append(self.map[0][x])
N = 60
m = maze(N,N)
def kruskal(Maze):
    outx = np.zeros((Maze.h,Maze.w),np.uint8)
    outy = np.zeros((Maze.h,Maze.w),np.uint8)
    outid = np.array(Maze.ids)
    check = outid == -1
    while not np.all(check):
        for c in Maze.cells:
            avail = []
            for i in c.neigs:
                if i.id != c.id:
                    avail.append(i)
            if len(avail) > 0:
                s = random.choice(avail)
                y,x = c.connect(s)
                outx[c.y][c.x] = (x + 3*abs(x))//2 + (y + 7*abs(y))//2
                outy[c.y][c.x] = (y + 3*abs(y))
            outid[c.y][c.x] = c.id
            for p in Maze.cells:
                an = p.ancestor
                while(an != an.ancestor):
                    an = an.ancestor
                p.ancestor = an
                p.id = an.id
        check = outid == outid[0][0]
        #print(check)
    return outy,outx,outid
_,out,_ = kruskal(m)
st = np.array([['█' for i in range(2*N+1)] for j in range(2*N+1)])
for i in range(N):
    for j in range(N):
        a = out[i][j]
        st[2*i+1][2*j+1] = ' '
        if a == 1:
            st[2*i+1][2*j] = ' '
        if a == 2:
            st[2*i+1][2*j+2] = ' '
        if a == 3:
            st[2*i][2*j+1] = ' '
        if a == 4:
            st[2*i+2][2*j+1] = ' '
mask = st[0]!=st[-1]
st[0][mask] = ' '
st[-1][mask] = ' ' 
mask2=st[:,0]!=st[:,-1]
st[mask2,0] = ' '
st[mask2,-1] = ' '
for i in st:
    for j in i:
        print(j,end='')
    print()
                    
        
```
